```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
# Using value counts method to make an array of all SN
unique_players = purchase_data['SN'].value_counts()
# Finding the length of unique_players array to put it in a dataframe under the column name "total players"
totalplayers = pd.DataFrame({"Total Players": [len(unique_players)]})
totalplayers.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#unique items count stored as a dataframe

# using length of the value count array of item name to find the unique items
unique_items_count = len(purchase_data["Item Name"].value_counts())
# to find avg purchase price, run the mean function on the price column and round value to 2 decimals places
average_purchase_price = round(purchase_data["Price"].mean(),2)
# count of purchases should just be the length of the price column
total_purchases = len(purchase_data["Price"])
# using sum function of price column
total_revenue = purchase_data["Price"].sum()
# creating a dataframe using a dictionary of lists that you defined
purchasing_analysis = pd.DataFrame({"Unique Items Count": [unique_items_count], "Average PurchasePrice": [average_purchase_price],
                                    "Total Purchases": [total_purchases],
                                    "Total Revenue": [total_revenue]
                                   })
purchasing_analysis.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# creating a dataframe using the SN and Gender columns from orginal dataframe
genders = pd.DataFrame(purchase_data.loc[:,['SN','Gender']])
# droping the duplicate SN to find unique SN and associated Gender
unique_genders = genders.drop_duplicates(subset = 'SN')
# creating a dataframe of gender column with unique value counts
unique_genders = pd.DataFrame(unique_genders['Gender'].value_counts())
# taking the sum of the total count to use in percentage forloop function
total_people = unique_genders['Gender'].sum()
# list comprehension to iterate through unique genders and divide each value by the sum
unique_genders["Percentage of Players"] = [round(gender/total_people *100 ,2) for gender in unique_genders['Gender']]
# renaming Gender column to total count
unique_genders = unique_genders.rename(columns = {"Gender": "Total Count"})
unique_genders.index.name = "Gender"

unique_genders.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchases_gender = purchase_data.loc[:,['Gender']]
purchases_gender = pd.DataFrame(purchase_data['Gender'].value_counts())

grouped_df = purchase_data.groupby(["Gender"])
average_price = grouped_df["Price"].mean()
total_purchase_value = grouped_df["Price"].sum()

purchases_gender = purchases_gender.rename(columns={"Gender": "Purchase Count"})
purchases_gender.index.name = "Gender"
purchases_gender["Average Purchase Price"] = average_price
purchases_gender["Total Purchase Value"] = total_purchase_value
purchases_gender["Avg Total Purchase per Person"] = [a/b for a, b in  zip(purchases_gender["Total Purchase Value"],
                                                                          unique_genders['Total Count'])]


purchases_gender["Average Purchase Price"] = purchases_gender["Average Purchase Price"].map("${:.2f}".format)
purchases_gender["Total Purchase Value"] = purchases_gender["Total Purchase Value"].map("${:,.2f}".format)
purchases_gender["Avg Total Purchase per Person"] = purchases_gender["Avg Total Purchase per Person"].map("${:.2f}".format)


# print(average_price)
# print(total_purchase_value)
# print(total_purchase_list)
# print(average_purchase_list)
purchases_gender.head()


```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 120]
group_names = ["<10", "10-14","15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

age_demo = purchase_data.loc[:,["Age", "SN"]]
age_demo = pd.DataFrame(age_demo)
age_demo = age_demo.drop_duplicates(["SN"])
del age_demo['SN']

age_demo["Age Range"] = pd.cut(age_demo["Age"], bins, labels=group_names)
age_demo["Age Range"].value_counts()
grouped_age = age_demo.groupby(["Age Range"])

age_count = grouped_age.count()
age_count = age_count.rename(columns={"Age": "Total Count"})
age_count_list = list(age_count["Total Count"])
age_percent_list = [round((age/sum(age_count_list))*100,2) for age in age_count_list]
age_count["Percentage of Players"] = age_percent_list
age_count.head(10)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 120]
group_names = ["<10", "10-14","15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

age_purchases = purchase_data.loc[:,["Age", "Price"]]
age_purchases["Age Range"] = pd.cut(age_purchases["Age"], bins, labels=group_names)
grouped_age_purchases = age_purchases.groupby(["Age Range"])
grouped_average_purchases = grouped_age_purchases["Price"].mean().map("${:.2f}".format)
grouped_total_purchases = grouped_age_purchases["Price"].sum()

age_purchases = grouped_age_purchases.count()
age_purchases = age_purchases.rename(columns={"Age": "Purchase Count"})
age_purchases["Average Purchase Price"] = grouped_average_purchases
age_purchases["Total Purchase Value"] = grouped_total_purchases
avg_total_purchase = [a/b for a,b in zip(age_purchases["Total Purchase Value"], age_count["Total Count"])]

age_purchases["Avg Total Purchase per Person"] = avg_total_purchase
age_purchases["Avg Total Purchase per Person"] = age_purchases["Avg Total Purchase per Person"].map("${:.2f}".format)

age_purchases["Total Purchase Value"] = grouped_total_purchases.map("${:,.2f}".format)


del age_purchases["Price"]
age_purchases.head(8)
```

## Top Spenders

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_df = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]


grouped_pop = popular_df.groupby(["Item ID", "Item Name"])
grouped_pop_count = grouped_pop.count()
grouped_pop_total = grouped_pop.sum()



pop_item_price = round(grouped_pop_total/grouped_pop_count,2)



popular_df = grouped_pop_count
popular_df["Item Price"] = pop_item_price
popular_df["Total Purchase Value"] = grouped_pop_total

popular_df["Total Purchase Value"]= popular_df["Total Purchase Value"].map("${:,.2f}".format)
popular_df["Item Price"] = popular_df["Item Price"].map("${:.2f}".format)
popular_df = popular_df.sort_values("Price", ascending = False)
popular_df = popular_df.rename(columns={"Price":"Purchase Count"})



# grouped_pop.head()
# grouped_pop_total
popular_df.head()
```
